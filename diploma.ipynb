```python
from sklearn.datasets import load_boston
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import pandas as pd
#Load boston housing dataset as an example
dataset=pd.read_csv("../input/dataset50/DataSet50.csv", ';')
#dataset=pd.read_csv("../input/testdata/testData.csv", ';')
dataset = pd.get_dummies(dataset, columns=['type']) 
dataset = pd.get_dummies(dataset, columns=['nameService']) 
features = dataset.drop('Class', 1) 
labels = dataset['Class']
names = ['sessionId', 'accountBalance', 'accountLimit', 'transactionAmount', 'commissionAmount', 'IDDevice', 'anomalTransfer', 'NewPayee', 'regular', 'gray', 'atypicalGeo', 'deviceConf','time', 'type_SERVICE-PAYMENT', 'type_TRANSFER-BETWEEN-ACCOUNTS', 'type_TRANSFER-IS-ACCOUNT', 'type_TRANSFER-IS-CARD', 'type_TRANSFER-TO-THE-BUDGET', 'nameService_IE', 'nameService_LLC', 'nameService_PE', 'nameService_SE', 'nameService_internal', 'nameService_internalBank', 'nameService_phone', 'nameService_service']
rf = RandomForestRegressor(n_estimators = 200,criterion='mse', min_samples_split=2, criterion='gini', max_features='auto')
rf.fit(features, labels)
print (sorted(zip(map(lambda x: round(x, 4), rf.feature_importances_), names), reverse=True))
```

```python
from sklearn.datasets import load_boston
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import pandas as pd

dataset=pd.read_csv('../input/DataSet3500.csv', ';')

dataset = pd.get_dummies(dataset, columns=['type']) 
dataset = pd.get_dummies(dataset, columns=['nameService']) 
features = dataset.drop('Class', 1) 
labels = dataset['Class']
names = ['deviceConf', 'gray', 'anomalTransfer', 
         'NewPayee', 'sessionId', 'nameService_PE', 
         'atypicalGeo','time', 'commissionAmount', 
         'transactionAmount', 'accountBalance', 'accountLimit',
         'nameService_phone', 'type_SERVICE-PAYMENT', 'regular', 
         'type_TRANSFER-IS-CARD', 'nameService_IE', 
         'nameService_service', 'IDDevice', 'type_TRANSFER-IS-ACCOUNT', 
         'nameService_internal', 'type_TRANSFER-BETWEEN-ACCOUNTS',
         'nameService_internalBank', 'type_TRANSFER-TO-THE-BUDGET'
         ,'nameService_LLC','nameService_SE' ]
rf = RandomForestClassifier(n_estimators = 200,criterion='gini',
                            min_samples_split=2, max_features='sqrt')
rf.fit(features, labels)
print (sorted(zip(map(lambda x: round(x, 4), 
                      rf.feature_importances_), names), reverse=True))
fs=sorted(rf.feature_importances_, reverse=True)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colorbar import Colorbar

names =['accountLimit','atypicalGeo', 'nameService_internal', 'time',  'deviceConf','transactionAmount', 'type_SERVICE-PAYMENT', 'nameService_LLC', 'sessionId', 'nameService_SE', 'NewPayee', 'anomalTransfer', 'gray', 'type_TRANSFER-IS-CARD', 'accountBalance', 'nameService_IE', 'nameService_internalBank', 'commissionAmount', 'nameService_phone', 'type_TRANSFER-TO-THE-BUDGET', 'type_TRANSFER-IS-ACCOUNT', 'IDDevice', 'nameService_service', 'type_TRANSFER-BETWEEN-ACCOUNTS', 'regular', 'nameService_PE']
fig, axes = plt.subplots(1, 1, figsize=(20, 8), dpi=300);
f1=axes.bar(range(0, 26),fs,edgecolor='darkgreen', alpha=1, zorder=2)
axes.grid(lw=0.25, color='darkgreen', zorder=1)
axes.set_xlabel('Признаки', fontsize=19)
axes.set_ylabel('Влияние', fontsize=19)
axes.set_title('Влияние признаков на обучение', fontsize=22, fontweight='bold')
axes.set_xticklabels(labels=names,rotation=90)
axes.set_xticks(np.arange(min(fs), max(fs)+25, 1))
axes.tick_params(axis='both', which='major', labelsize=16)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, f1_score, fbeta_score
from sklearn.model_selection import cross_validate, cross_val_score, cross_val_predict

scoresRF_f1 = []
scoresRF_auc = []

from sklearn.model_selection import KFold
kf = KFold(n_splits=5)

def auc(y_true, y_scores): return roc_auc_score(y_true, y_scores,average='macro')
def fb(y_true, y_scores): return fbeta_score(y_true, y_scores,average='macro', beta=2)

from sklearn.metrics import make_scorer
scoring = {'f1_macro':make_scorer(fb), 'auc': make_scorer(auc)}

for k in range(10, 351, 10):
    RF = RandomForestClassifier(n_estimators=k, max_depth=None,random_state=0, criterion='entropy', max_features='sqrt',n_jobs=-1)
    scoreRF=cross_validate(RF, featuresNormMM, labels, cv=kf, scoring=scoring, return_train_score=False)
    scoresRF_f1.append(scoreRF['test_f1_macro'].mean())
    scoresRF_auc.append(scoreRF['test_auc'].mean())
```

```python
df = pd.get_dummies(df, columns=['type'])
X=df.drop(['isFraud'],1)
y=df['isFraud']

from sklearn.model_selection import train_test_split 
features_train, features_test, labels_train, labels_test = train_test_split(X, y, test_size=0.2, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
modelDTC = DecisionTreeClassifier(criterion= 'entropy', splitter='best',max_depth = None, max_features = 'sqrt')
modelDTC.fit(features_train, labels_train)
#print(modelDTC)

expected = labels_train
predicted = modelDTC.predict(features_train)

print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```

```python
expected = labels_test
predicted = modelDTC.predict(features_test)

print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```
